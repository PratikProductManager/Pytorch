```python
import torch
import torch.nn as nn
from torch.nn import functional as F

import pandas as pd
import numpy as np
```

```python
colab_scaler= pickle.load(open('sc.pickle.1','rb'))
```

```python
input_size=2
output_size=2
hidden_size=10
```

```python
class Net(nn.Module):
   def __init__(self):
       super(Net, self).__init__()
       self.fc1 = torch.nn.Linear(input_size, hidden_size)
       self.fc2 = torch.nn.Linear(hidden_size, hidden_size)
       self.fc3 = torch.nn.Linear(hidden_size, output_size)


   def forward(self, X):
       X = torch.relu((self.fc1(X)))
       X = torch.relu((self.fc2(X)))
       X = self.fc3(X)

       return F.log_softmax(X,dim=1)
```

```python
new_predictor2=Net()
```

```python
new_predictor2.load_state_dict(torch.load('customer_buy_state_dict.1',weights_only=False))
```

```python
y_cust = new_predictor2(torch.from_numpy(colab_scaler.transform(np.array([[20,20000]]))).float())
y_cust
```
